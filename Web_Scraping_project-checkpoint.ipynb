#### Multiple pages

```python
import requests 
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
Hotel_names=[]
Price=[]
Location=[]
Distance=[]
Star_rating=[]
Original_Price=[]
Discount_off=[]
Ratedby=[]
Facilities=[]

for i in range(2,9):
    url="https://www.oyorooms.com/hotels-in-hyderabad/?page={i}"
    page=requests.get(url)
    page=requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    content=page.text
    internet=BeautifulSoup(content)

    
    # hotel name
    hotel = internet.find_all("div",class_="listingHotelDescription__contentWrapper")

    for i in hotel:
        a=i.find("a",class_="c-nn640c u-width100")                
        text=a.text
        Hotel_names.append(text)

    # price        
    price = internet.find_all("span",class_="listingPrice__finalPrice")
    
    for i in price:
        a=i.text
        b=re.findall(r"\d+",a)
        Price.append(b[0])
    
    # location
    location = internet.find_all("span",class_="u-line--clamp-2")

    for i in hotel:
        a=i.find("span",class_="u-line--clamp-2")
        b=a.text
    
        regex=re.findall("(.*),",b)
    
        if regex:
            Location.append(regex[0])
        else:
            Location.append(np.nan)
        
    # distance    
    distance = internet.find_all("span",class_="listingHotelDescription__distanceText")    
    
    for i in distance:
        a=i.text
        b=re.findall(r"(\d+\.\d+)\skm",a)
        if len(b)>0:
            Distance.append(b[0])
    
    # star rating
    rating = internet.find_all("div",class_="hotelRating")

    for i in rating:
        Star_rating.append(i.text.split("(")[0])
        

    # original price
    original_price = internet.find_all("span",class_="listingPrice__slashedPrice d-body-lg")
    
    for i in original_price:
        a=i.text
        b=re.findall(r"\d+",a)
        Original_Price.append(b[0])
        
    # discount
    discount = internet.find_all("span",class_="listingPrice__percentage")
    
    for i in discount:
        a=i.text
        b=re.findall(r"\d+",a)[0]
        if len(b)>0:
            Discount_off.append(b)
    # rated by    
    rating = internet.find_all("div",class_="hotelRating")

    for i in rating:
        a=i.find("span",class_="hotelRating__ratingSummary")
        b=i.text
        regex=re.findall("\((\d+\s)",b)
        if regex:
            Ratedby.append(regex[0])
        else:
            Ratedby.append(np.nan)
    
    # facility
    facility = internet.find_all("div",class_="amenityWrapper")

    for i in facility:
        Facilities.append(i.text)
```

```python
data1 = {
    "Hotel_names":Hotel_names,
    "Final_price(₹)":Price,
    "Location":Location,
    "Km":Distance,
    "Star_rating":Star_rating,
    "Original_price(₹)":Original_Price,
    "Discount_in(%)":Discount_off,
    "Ratedby":Ratedby,
    "Facility":Facilities
}
```

```python
oyo_hyderabad = pd.DataFrame(data1)
```

```python
oyo_hyderabad["City"]=("Hyderabad")
```

```python
oyo_hyderabad
```

```python
oyo_hyderabad.to_csv("result1.csv")
```
